```python
import numpy as np
import pandas as pd
import re
import openpyxl
```

```python
infile = open('airport-data-dictionaries.xlsx', mode='r+b')
```

```python
df = pd.read_excel(infile, sheet_name=0, index_col='Field', usecols='A,C')

pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 110)
pd.set_option('display.max_colwidth', None)
df
```

```python
df.index.rename('category', inplace=True)
df.head(2)
```

```python
df.columns = ['description']
df.head(0)
```

```python
#NOTE: 'ActivationDate' was modified to 'ActiviationDate' to match typo in source file. 

unused_labels = ['DistrictOffice', 'CountyState', 'ARPMethod', 'ARPElevationMethod', 'LandAreaCoveredByAirport', 'BoundaryARTCCComputerID',
                'ResponsibleARTCCComputerID', 'AirportToFSSPhoneNumber', 'AlternateFSSID', 'AlternateFSSName',
                'AlternateFSSTollFreeNumber', 'ActiviationDate', 'CertificationTypeDate', 'FederalAgreements','AirspaceDetermination',
                'InspectionMethod', 'InspectionGroup', 'LastInspectionDate', 'LastOwnerInformationDate', 'SegmentedCircle', 'AirportPositionSource',
                'AirportPositionSourceDate', 'AirportElevationSource', 'AirportElevationSourceDate', 'ContractFuelAvailable', 'TransientStorage',
                'OtherServices', 'WindIndicator', 'MinimumOperationalNetwork']
```

```python
df.drop(index=unused_labels, inplace=True)
df.head(10)
```

```python
labels = df.index
```

```python
df.index = [re.sub(r'(?<=[a-z])(?=[A-Z])|(?<=[A-Z])(?=[A-Z][a-z])|(?<=[A-Z])(?=ID)', '_', x) 
            for x in labels]
```

```python
df.index = df.index.str.strip().str.lower()
```

```python
df.head()
```

```python
infile.close()
```

```python
outfile = '../Clean-Data/data_dictionary.xlsx'
```

```python
with open(outfile, mode='wb'):
    wb = openpyxl.Workbook()
    wb.save(filename=outfile)
    
    df.to_excel(outfile, sheet_name='Facilities', index_label='category')
```
