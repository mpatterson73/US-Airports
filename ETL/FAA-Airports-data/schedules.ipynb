```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv('orig_schedules_faa.tsv', sep='\t+', index_col=False,  engine='python')
pd.set_option('display.max_columns', 10)
pd.set_option('display.max_rows', 1670)
pd.set_option('display.max_colwidth', None)
df.head(15)
```

```python
df.shape
```

```python
df.head()
```

```python
df.columns = [re.sub(r'(?<=[a-z])(?=[A-Z])|(?<=[A-Z])(?=[A-Z][a-z])|(?<=[A-Z])(?=ID)', '_', x) 
              for x in df.columns]
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace('"', '')
```

```python
df.rename(columns={'attendanace_schedule': 'service_schedule'}, inplace=True)
```
