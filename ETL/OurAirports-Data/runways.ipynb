```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Airports/runways.csv')
pd.set_option('display.max_rows', 200)
df.head()
```

```python
adf = pd.read_csv('Clean-Data/airports_clean.csv')
df = df[df['airport_ident'].isin(adf['ident'])]
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['id', 'airport_ref', 'le_heading_degT', 'he_heading_degT'], inplace=True)
```

```python
df.columns = ['airport_ident', 'length', 'width', 'surface', 'lighted',
       'closed', 'low_end', 'le_lat', 'le_long',
       'le_elev', 'le_threshold', 'high_end',
       'he_lat', 'he_long', 'he_elev',
       'he_threshold']
```

```python
df[['lighted', 'closed']].apply(pd.Series.value_counts)
```

```python
df['lighted'] = df['lighted'].map({1: True, 0: False})
df['closed'] = df['closed'].map({1: True, 0: False})
df.head()
```
