```python
import numpy as np
import pandas as pd
import re
```

```python
cols_to_keep = ['SiteNumber', 'State', 'RunwayID', 'RunwayLength', 'RunwayWidth', 'RunwaySurfaceTypeCondition', 'RunwaySurfaceTreatment', 'EdgeLightsIntensity', 'BaseEndID', 'BaseEndTrueAlignment', 'BaseEndILSType', 'BaseEndRightTrafficPattern', 'BaseEndMarkingsType', 'BaseEndMarkingsCondition', 'BaseEndPhysicalElevation', 'BaseEndCrossingHeight', 'BaseEndGlidePathAngle', 'BaseEndDisplacedLength', 'BaseEndVASI', 'BaseEndCenterlineLights', 'BaseEndTouchdownLights', 'BaseEndObjectDescription', 'BaseEndObjectHeight', 'ReciprocalEndID', 'ReciprocalEndTrueAlignment', 'ReciprocalEndILSType', 'ReciprocalEndRightTrafficPattern', 'ReciprocalEndMarkingsType', 'ReciprocalEndMarkingsCondition', 'ReciprocalEndPhysicalElevation', 'ReciprocalEndCrossingHeight', 'ReciprocalEndGlidePathAngle', 'ReciprocalEndDisplacedLength', 'ReciprocalEndVASI', 'ReciprocalEndCenterlineLights', 'ReciprocalEndTouchdownLights', 'ReciprocalEndObjectDescription', 'ReciprocalEndObjectHeight', 'RunwayWeightBearingCapacitySW', 'RunwayWeightBearingCapacityDW', 'RunwayWeightBearingCapacityDT', 'RunwayWeightBearingCapacityDDT', 'BaseEndTakeOffRunAvailableTORA', 'BaseEndTakeOffDistanceAvailableTODA', 'BaseEndAcltStopDistanceAvailableASDA', 'BaseEndLandingDistanceAvailableLDA', 'ReciprocalEndTakeOffRunAvailableTORA', 'ReciprocalEndTakeOffDistanceAvailableTODA', 'ReciprocalEndAcltStopDistanceAvailableASDA', 'ReciprocalEndLandingDistanceAvailableLDA']
```

```python
df = pd.read_csv('../Raw-FAA-Data/orig_runways_faa.tsv', sep='\t', index_col=False, usecols=cols_to_keep, engine='python')
pd.set_option('display.max_columns', 140)
pd.set_option('display.max_rows', 110)
pd.set_option('display.max_seq_items', 175)
df.head()
```

```python
df.shape
```

```python
col_list = df.columns
```

```python
df.columns = [re.sub(r'(?<=[a-z])(?=[A-Z])|(?<=[A-Z])(?=[A-Z][a-z])|(?<=[A-Z])(?=ID)', '_', x) 
              for x in col_list]
```

```python
df.columns = df.columns.str.strip().str.lower()
```

```python
df.columns
```

```python
#col_list = df.columns
df.columns = [re.sub(r'_end|(_s$)', '', col) for col in df.columns]
df.columns = [re.sub(r'reciprocal', 'rcpr', col) for col in df.columns]
df.columns = [re.sub(r'rcpr_id', 'reciprocal_end_id', col) for col in df.columns]
df.columns = [re.sub(r'base_id', 'base_end_id', col) for col in df.columns]
df.columns = [re.sub(r'condition', 'cond', col) for col in df.columns]
df.columns = [re.sub(r'_physical|_description|_traffic|', '', col) for col in df.columns]
df.columns = [re.sub(r'weight_bearing_capacity', 'capacity_lbs', col) for col in df.columns]
df.columns = [re.sub(r'_take_off_run_available|_take_off_distance_available|_aclt_stop_distance_available|_landing_distance_available',
                    '', col) for col in df.columns]
```

```python
df.columns
```

```python
df[['base_end_id', 'reciprocal_end_id']] = df[['base_end_id', 'reciprocal_end_id']].applymap(lambda x: x[1:])
df.head()
```

```python
df.info()
```

```python
bool_cols = ['base_right_pattern', 'base_centerline_lights', 'base_touchdown_lights', 'rcpr_right_pattern', 'rcpr_centerline_lights', 'rcpr_touchdown_lights']
```

```python
for col in bool_cols:
    df[col] = df[col].map({'Y': True, 'N': False})
```

```python
values = {'runway_surface_type_cond': 'UNKNOWN', 'runway_surface_treatment': 'NONE', 'edge_lights_intensity': 'UNKNOWN',
         'base_ils_type': 'NONE', 'base_right_pattern': False, 'base_markings_type': 'NONE', 'base_markings_cond': 'UNKNOWN',
         'base_latitude': 'NONE', 'base_longitude': 'NONE', 'base_vasi': 'NONE', 'base_centerline_lights': False,
         'base_touchdown_lights': False, 'base_object': 'NONE', 'rcpr_ils_type': 'NONE', 'rcpr_right_pattern': False, 'rcpr_markings_type': 'NONE', 'rcpr_markings_cond': 'UNKNOWN',
         'rcpr_latitude': 'NONE', 'rcpr_longitude': 'NONE', 'rcpr_vasi': 'NONE', 'rcpr_centerline_lights': False,
         'rcpr_touchdown_lights': False, 'rcpr_object': 'NONE', }
```

```python
df.fillna(value=values, inplace=True)
```
